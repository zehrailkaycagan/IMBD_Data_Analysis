```python
def rating(rating):
    if rating>=7.0:
        return "Excelent"
    elif rating>=6.0:
        return"Good"
    else:
        return"Average"
```

```python
#bi önceki kodda verilen rating sutunundaki herbir değeri rating fonksiyonuna göndeririz.Bu fonksiyondan dönen değeri rating_cat sütununa ekler ve kategorik değerler içerir
data['rating_cat']=data['Rating'].apply(rating)
```
